```python
import pandas as pd
import numpy as np
```

## Combine texts

```python
df = pd.read_csv("new_new_bolt.csv")
```

```python
df['helper'] = (df['speaker'] != df['speaker'].shift()) | (df['convo_id'] != df['convo_id'].shift())
df['helper_cumsum'] = df['helper'].cumsum()

df = df.groupby(['helper_cumsum', 'speaker', 'convo_id'], as_index=False)['text'].agg(', '.join)
df = df.drop('helper_cumsum', axis=1)
```

```python
df.to_csv("combined.csv", index=False)
```

## Pred DF

```python
text_df = pd.read_csv("combined.csv")
```

```python
pred_df = pd.DataFrame(columns=['text', 'speaker', 'conversation'] + list(range(11)))
pred_df['text'] = text_df['text']
pred_df['speaker'] = text_df['speaker']  
pred_df['conversation'] = text_df['convo_id']
```

```python
pred_df = pred_df.dropna(subset=['text'])
```
